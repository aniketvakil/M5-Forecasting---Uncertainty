```python
submission1 = pd.read_csv('../input/top-submmm/submission.csv')
submission2 = pd.read_csv('../input/top-submmm/submissionUncertaintyDay1500.csv')
submission3 = pd.read_csv('../input/top-submmm/submission (1).csv')
submission4 = pd.read_csv('../input/top-submmm/submission (2).csv')
submission5 = pd.read_csv('../input/top-submmm/submission (3) (1).csv')
submission6 = pd.read_csv('../input/top-submmm/submission (3).csv')
submission7 = pd.read_csv('../input/top-submmm/submission (4).csv')
submission8 = pd.read_csv('../input/top-submmm/submission (5).csv')
submission9 = pd.read_csv('../input/top-submmm/submission (6).csv')
submission10 = pd.read_csv('../input/top-submmm/submission (7).csv')
submission11 = pd.read_csv('../input/top-submmm/submission (8).csv')
submission12 = pd.read_csv('../input/top-submmm/submissionUncertainty.csv')
#submission13 = pd.read_csv('../input/top-submmm/submissionUncertainty1.csv')
```

```python
submission1 = submission1.sort_values(by = 'id').reset_index(drop = True)
submission2  = submission2.sort_values(by = 'id').reset_index(drop = True)
submission3 = submission3.sort_values(by = 'id').reset_index(drop = True)
submission4 = submission1.sort_values(by = 'id').reset_index(drop = True)
submission5  = submission2.sort_values(by = 'id').reset_index(drop = True)
submission6 = submission3.sort_values(by = 'id').reset_index(drop = True)
submission7 = submission1.sort_values(by = 'id').reset_index(drop = True)
submission8  = submission2.sort_values(by = 'id').reset_index(drop = True)
submission9 = submission3.sort_values(by = 'id').reset_index(drop = True)
submission10 = submission1.sort_values(by = 'id').reset_index(drop = True)
submission11  = submission2.sort_values(by = 'id').reset_index(drop = True)
submission12 = submission3.sort_values(by = 'id').reset_index(drop = True)

sub5 = submission1.copy()

for i in sub5.columns :
    if i != 'id' :
        sub5[i] = 0.1*submission1[i] + 0.1*submission2[i]+0.1*submission3[i]+0.1*submission4[i] + 0.1*submission5[i]+0.1*submission6[i]+0.05*submission7[i] + 0.05*submission8[i]+0.05*submission9[i]+0.05*submission10[i] + 0.1*submission11[i]+0.1*submission12[i]
        
sub5.to_csv('submission.csv', index = False)
```
